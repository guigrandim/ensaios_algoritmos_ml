# 0.0 Imports

```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics as mt
```

# 0.1 - Load Datasets

```python
actual_folder = Path.cwd()
print(actual_folder)

main_folder = actual_folder.parent
print(main_folder)

path_folder_dataset = main_folder / 'dataset' / 'classification_datasets'

if path_folder_dataset.exists():
    print(f'Caminho existente: {path_folder_dataset}')
else:
    print(f'Erro: O arquivo não foi encontrado {path_folder_dataset}')
```

```python
#Carrega os datasets da pasta - Treinamento / Validação e Teste

dataset_traning_X = path_folder_dataset / 'a_traninig' / 'X_training.csv'
dataset_traning_y = path_folder_dataset / 'a_traninig' / 'y_training.csv'

dataset_validation_X = path_folder_dataset / 'b_validation' / 'X_validation.csv'
dataset_validation_y = path_folder_dataset / 'b_validation' / 'y_validation.csv'

dataset_test_X = path_folder_dataset / 'c_test' / 'X_test.csv'
dataset_test_y = path_folder_dataset / 'c_test' / 'y_test.csv'

X_train = pd.read_csv(dataset_traning_X)
y_train = pd.read_csv(dataset_traning_y)

X_val = pd.read_csv(dataset_validation_X)
y_val = pd.read_csv(dataset_validation_y)

X_test = pd.read_csv(dataset_test_X)
y_test = pd.read_csv(dataset_test_y)
```

# 1.0 - Training Algorithm

## Passo 2 — Treino com parâmetros default

```python
# Instanciar o modelo com parâmetros default
model_def = RandomForestClassifier(random_state=42)

# Treinar com dados de treino
model_def.fit(X_train, y_train.values.ravel())

# Predições no treino e na validação
yhat_train_def = model_def.predict(X_train)
yhat_val_def   = model_def.predict(X_val)
```

## Passo 3 — Performance no treino (default)

```python
# Métricas no conjunto de TREINO com parâmetros default
acuracia_train_def  = mt.accuracy_score(y_train, yhat_train_def)
precision_train_def = mt.precision_score(y_train, yhat_train_def)
recall_train_def    = mt.recall_score(y_train, yhat_train_def)
f1score_train_def   = mt.f1_score(y_train, yhat_train_def)

print("--- Performance no Treino (Default) ---")
print(f"Acurácia:  {acuracia_train_def:.4f}")
print(f"Precisão:  {precision_train_def:.4f}")
print(f"Recall:    {recall_train_def:.4f}")
print(f"F1-Score:  {f1score_train_def:.4f}")
```

## Passo 4 — Performance na validação (default)

```python
# Métricas no conjunto de VALIDAÇÃO com parâmetros default
acuracia_val_def  = mt.accuracy_score(y_val, yhat_val_def)
precision_val_def = mt.precision_score(y_val, yhat_val_def)
recall_val_def    = mt.recall_score(y_val, yhat_val_def)
f1score_val_def   = mt.f1_score(y_val, yhat_val_def)

print("--- Performance na Validação (Default) ---")
print(f"Acurácia:  {acuracia_val_def:.4f}")
print(f"Precisão:  {precision_val_def:.4f}")
print(f"Recall:    {recall_val_def:.4f}")
print(f"F1-Score:  {f1score_val_def:.4f}")
```

## Passo 5 — Ajuste de hiperparâmetros

```python
print("--- Testando Hiperparâmetros na Validação ---")

melhor_max_depth = 5
melhor_f1score_val = 0

list_test = list(range(1, 20, 1))
list_max_depth = list_test + [None]

for md in list_max_depth:
    #Treinamento apenas com o bloco de Treino - Define
    model = RandomForestClassifier( n_estimators=100,
                                    max_depth=md,
                                    random_state=42)
    
    #Traning
    model.fit (X_train, y_train.values.ravel())
    
    #Predict
    yhat_train = model.predict(X_train)
    yhat_val = model.predict(X_val)
    
    #Performance Training
    accuracia_train = mt.accuracy_score(y_train, yhat_train)
    precisao_train = mt.precision_score(y_train, yhat_train)
    recall_train = mt.recall_score(y_train, yhat_train)
    f1score_train = mt.f1_score(y_train, yhat_train)
    
    #Performance Validation
    accuracia_val = mt.accuracy_score(y_val, yhat_val)
    precisao_val = mt.precision_score(y_val, yhat_val)
    recall_val = mt.recall_score(y_val, yhat_val)
    f1score_val = mt.f1_score(y_val, yhat_val)
    
    # Rótulo visual para quando o k for None
    label_md = "None" if md is None else f"{md:4d}"
    
    # Converte o md para string primeiro ('2', '5', 'None') para aceitar formatação de texto (:>4)
    label_md = str(md)
    
    print(f"K = {label_md:>4} | Acurácia Treino: {accuracia_train:.4f}     | Acurácia Validação:   {accuracia_val:.4f}")
    print(f"K = {label_md:>4} | Precisão Treino: {precisao_train:.4f}      | Precisão Validação:   {precisao_val:.4f}")
    print(f"K = {label_md:>4} | Recall Treino:   {recall_train:.4f}        | Recall Validação:     {recall_val:.4f}")
    print(f"K = {label_md:>4} | F1-Score Treino: {f1score_train:.4f}       | F1-Score Validação:   {f1score_val:.4f}")
    print("=" * 60)
    
    # Salva o melhor K com base na validação
    if f1score_val > melhor_f1score_val:
        melhor_f1score_val = f1score_val
        melhor_max_depth = md

print(f"\n-> O melhor hiperparâmetro escolhido foi: max_depth={melhor_max_depth}\n")
    
```

```python
print("--- Testando Múltiplos Hiperparâmetros no Random Forest ---")
melhor_max_depth = 5
melhor_n_estimators = 100
melhor_f1score_val = 0

# Suas duas listas prontas para o ensaio
list_test = list(range(1, 20, 1))
list_max_depth = list_test + [None]
list_n_estimator = [50, 60, 70, 80, 90, 100]

# LOOP DUPLO: Cruza cada profundidade com cada quantidade de árvores
for md in list_max_depth:
    for ne in list_n_estimator:
        
        # Passando os dois parâmetros variáveis ao mesmo tempo
        model = RandomForestClassifier(
            n_estimators=ne,
            max_depth=md,
            min_samples_split=10, 
            min_samples_leaf=5, 
            random_state=42,
            n_jobs=-1
        )
        
        # Training
        model.fit(X_train, y_train.values.ravel())
        
        # Predict
        yhat_train = model.predict(X_train)
        yhat_val = model.predict(X_val)
        
        # F1-Scores (Métrica chave do laboratório)
        f1score_train = mt.f1_score(y_train, yhat_train)
        f1score_val = mt.f1_score(y_val, yhat_val)
        
        # Formatando o rótulo do max_depth para o print
        label_md = str(md)
        
        # Print resumido para não inundar o seu terminal com 120 blocos gigantes
        print(f"MD: {label_md:>4} | Estimators: {ne:3d} | F1 Treino: {f1score_train:.4f} | F1 Validação: {f1score_val:.4f}")
        
        # Salva se encontrar a melhor combinação de parâmetros baseada na validação
        if f1score_val > melhor_f1score_val:
            melhor_f1score_val = f1score_val
            melhor_max_depth = md
            melhor_n_estimators = ne

print("=" * 65)
print(f"-> VENCEDOR DO ENSAIO:")
print(f"Melhor max_depth: {melhor_max_depth}")
print(f"Melhor n_estimators: {melhor_n_estimators}")
print(f"Maior F1-Score obtido na Validação: {melhor_f1score_val:.4f}\n")
```

## Passo 6 — União treino + validação

```python
# Unir treino + validação para formar o conjunto final de treinamento
X_train_final = pd.concat([X_train, X_val])
y_train_final = pd.concat([y_train, y_val])

print(f"X_train_final shape: {X_train_final.shape}")
print(f"y_train_final shape: {y_train_final.shape}")
```

## Passo 7 — Retreino com melhores parâmetros

```python
n_estimators = 100
max_depth = 19

# Passando os dois parâmetros variáveis ao mesmo tempo
model = RandomForestClassifier( n_estimators=n_estimators,
                                max_depth=max_depth,
                                min_samples_split=10, 
                                min_samples_leaf=5, 
                                random_state=42,
                                n_jobs=-1
                            )
        
# Training
model.fit(X_train, y_train.values.ravel())
        
# Predict
yhat_train = model.predict(X_train)
yhat_val = model.predict(X_val)
        
# Métricas de Avaliação do Modelo (Métrica chave do laboratório)
acuracia_train = mt.accuracy_score(y_train, yhat_train)
acuracia_val = mt.accuracy_score(y_val, yhat_val)

precision_train = mt.precision_score(y_train, yhat_train)
precision_val = mt.precision_score(y_val, yhat_val)

recall_train = mt.recall_score(y_train, yhat_train)
recall_val = mt.recall_score(y_val, yhat_val)

f1score_train = mt.f1_score(y_train, yhat_train)
f1score_val = mt.f1_score(y_val, yhat_val)

# Formatando o rótulo do max_depth para o print
label_md = str(max_depth)


print(f"MD: {label_md:>4} | Estimators: {n_estimators:3d}     | Acuracia Treino: {acuracia_train:.4f}      | Acuracia Validação: {acuracia_val:.4f}")
print(f"MD: {label_md:>4} | Estimators: {n_estimators:3d}     | Precision Treino: {precision_train:.4f}    | Precision Validação: {precision_val:.4f}")
print(f"MD: {label_md:>4} | Estimators: {n_estimators:3d}     | Recall Treino: {recall_train:.4f}          | Recall Validação: {recall_val:.4f}")
print(f"MD: {label_md:>4} | Estimators: {n_estimators:3d}     | F1 Treino: {f1score_train:.4f}             | F1 Validação: {f1score_val:.4f}")
```

## Passo 8 — Performance no teste

```python
# ==============================================================================
# JUNTAR AS BASES E EXECUTAR O TESTE FINAL
# ==============================================================================
print("--- Executando o Teste Final ---")

melhor_max_depth_test = melhor_max_depth
melhor_n_estimator_test = melhor_n_estimators

# Juntando treino e validação para dar mais volume de dados ao KNN
X_train_final = pd.concat([X_train, X_val])
y_train_final = pd.concat([y_train, y_val])

# Treinando o modelo definitivo com o melhor K encontrado
model_final = RandomForestClassifier(n_estimators = melhor_n_estimators,
                                     max_depth = melhor_max_depth,
                                     random_state=42)
model_final.fit(X_train_final, y_train_final.values.ravel())

# Previsão final usando o dataset de teste (que ficou isolado o tempo todo)
yhat_test = model_final.predict(X_test)

#Performance Test
accuracia_test = mt.accuracy_score(y_test, yhat_test)
precisao_test = mt.precision_score(y_test, yhat_test)
recall_test = mt.recall_score(y_test, yhat_test)
f1score_test = mt.f1_score(y_test, yhat_test)

# Métrica final de produção
print(f"MD = {melhor_max_depth_test:2d} + N_Estimador = {melhor_n_estimators:2d} | Acurácia Teste: {accuracia_test:.4f}")
print(f"MD = {melhor_max_depth_test:2d} + N_Estimador = {melhor_n_estimators:2d} | Precisão Teste: {precisao_test:.4f}")
print(f"MD = {melhor_max_depth_test:2d} + N_Estimador = {melhor_n_estimators:2d} | Recall Teste:   {recall_test:.4f}")
print(f"MD = {melhor_max_depth_test:2d} + N_Estimador = {melhor_n_estimators:2d} | F1-Score Teste: {f1score_test:.4f}")
```
